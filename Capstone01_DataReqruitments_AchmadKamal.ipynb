```python
def exctractDataCsv(path):
    data1 = pd.read_csv(path)

    #indek dihapus dari tampilan
    blanked_index=[''] * len(data1)
    data1.index=blanked_index

    return data1

data1 = exctractDataCsv('/content/drive/My Drive/purwadhika_files/data_sources/data_reqruitment/data_requirements.csv')
data1.info()
```

# B.Fitur Transformation

## 1.Drop Kolom Unnamed: 0

```python
data1.drop('Unnamed: 0',axis=1,inplace=True) #menghapus kolom unnamed: 0 karena tdk diperlukan
data1.head()
```

## 2.Handle Null Values

```python
data1.isna().sum() #cek apakah ada nilai null.
```

```python
#hampir semua kolom memiliki null value, yang akan saya hapus hanya row yang company nya null, karena tidak valid, untuk kolom lainnya masih bisa diambil kesimpulan jika nilainya null.
data1.dropna(subset=['company'], inplace=True)
data1.isna().sum()
```

## 3.Ubah data type Dates menjadi datetime type dan seragamkan timezone

cek data type terlebih dahulu

```python
data1.dtypes
```

mengubah kolom dates dari semula dtypes object menjadi dtypes Datetime, dan menyeragamkan timezone menjadi UTC+0

```python
data1['dates'] = pd.to_datetime(data1['dates'], format='%Y-%m-%d %H:%M:%S%z', utc=True)
data1.info()
```

```python
data1.head()
```

## 4.Ubah salary estimate menjadi float dan pastikan nilainya seragam (currency dan satuannya)

```python
#cek apakah semua salary estimate dalam $
data1['salary_estimate'].str[0].unique()
```

berarti hasil di atas menunjukkn semua
salary estimate dalam $

```python
#cek salary estimate /hr dan /yr saja atau tidak
data1['salary_estimate'].str.extract(r' /(\w+) ')[0].unique()
```

hasil di atas menunjukkan salary estimate satuannya hr(hour), yr(year), dan mo (month)

```python
#replace comma menjadi ''
data1['salary_estimate'] = data1['salary_estimate'].str.replace(',','')
data1.head()
```

```python
#extract nilai salary
data1['salary_estimate'].str.extract(r'(\d+.\d+)')[0]
data1.head()
```

```python
#buat kolom faktor kali est untuk standarisasi nilai salary est
data1['faktorkali_salary_est'] = data1['salary_estimate'].str.extract(r' /(\w+) ')[0]
data1['faktorkali_salary_est'] = data1['faktorkali_salary_est'].str.replace('yr','1848') #asumsi dalam 1 tahun rata2 work hour di US adalah 2087 jam. sumbee : https://www.indeed.com/career-advice/career-development/how-many-work-hours-in-a-year
data1['faktorkali_salary_est'] = data1['faktorkali_salary_est'].str.replace('hr','1')
data1['faktorkali_salary_est'] = data1['faktorkali_salary_est'].str.replace('mo','154') #asumsi rata2 jam kerja per bulan berdasar sumber jam kerja tahunan di atas dibagi 12
data1['faktorkali_salary_est']

```

```python
#fillna faktor kali salary est dg nilai 0
data1['faktorkali_salary_est'] = data1['faktorkali_salary_est'].fillna(0)
data1['faktorkali_salary_est']
```

```python
#membuat kolom std_salary_est_perhour
data1['value_salary_est'] = data1['salary_estimate'].str.extract(r'(\d+.\d+)')[0] #mencari value float dari extract string.
data1['value_salary_est'] = data1['value_salary_est'].astype(float) #mengubah datatype menjadi float agar bisa dioperasikan dg math operation
data1['std_salary_est_perhour'] = data1['value_salary_est'] / data1['faktorkali_salary_est'].astype(int) #math operation untuk standardise salary per hour in USD $
data1
```

## 5.Melakukan labelling pada kolom company_revenue

```python
#cek apakah company revenue dalam dollar USD smua
data1['company_revenue'].str.extract(r'\((.*?)\)')[0].unique()
```

```python
data1['company_revenue'].unique()
```

dari hasil di atas , semua company revanue dalam USD.

```python
#labeling company_revenue menjadi grade yang dapat diurutkan atau dilakukan agregrate
data1.loc[data1['company_revenue'] == 'Less than $1 million (USD)', 'grade_company_revanue'] = 1
data1.loc[data1['company_revenue'] == '$1 to $5 million (USD)', 'grade_company_revanue'] = 2
data1.loc[data1['company_revenue'] == '$5 to $25 million (USD)', 'grade_company_revanue'] = 3
data1.loc[data1['company_revenue'] == '$25 to $100 million (USD)', 'grade_company_revanue'] = 4
data1.loc[data1['company_revenue'] == '$100 to $500 million (USD)', 'grade_company_revanue'] = 5
data1.loc[data1['company_revenue'] == '$500 million to $1 billion (USD)', 'grade_company_revanue'] = 6
data1.loc[data1['company_revenue'] == '$1 to $5 billion (USD)', 'grade_company_revanue'] = 7
data1.loc[data1['company_revenue'] == '$5 to $10 billion (USD)', 'grade_company_revanue'] = 8
data1.loc[data1['company_revenue'] == '$10+ billion (USD)', 'grade_company_revanue'] = 9
data1.loc[data1['company_revenue'] == 'Unknown / Non-Applicable', 'grade_company_revanue'] = 0
data1.loc[data1['company_revenue'].isnull(),  'grade_company_revanue'] = 0

data1.head()
```

```python
data1['grade_company_revanue'] = data1['grade_company_revanue'].astype('int32')
data1.info()
```
