#A. Fitur Extract

```python
import pandas as pd
import pytz
from datetime import datetime
import re #untuk regex findall
import glob
```

```python
#buat function untuk meng extract semua file dalam direktori tertentu
def exctractDataCsv(path):
    # mendapatkan list dari file di direktori
    listFiles = glob.glob(path)

    # membuat dataframe kosongan sbg tempat data yg digabung dimasukkan
    concatedData1 = pd.DataFrame()

    # looping setiap file di direktori tsb dan isinya di append (ditambahkan) di dataframe gabunga hasil concat
    for i in listFiles:
        data1 = pd.read_csv(i)
        concatedData1 = pd.concat([concatedData1, data1])

    #index tidak ditampilkan
    blanked_index=[''] * len(concatedData1)
    concatedData1.index=blanked_index

    #return data yang sudah di concate dan disembunyikan indexnya
    return concatedData1


#panggil function untuk extractDataCSV untuk semua file csv yang ada di direktori data_products
concatedData1 = exctractDataCsv('/content/drive/My Drive/purwadhika_files/data_sources/data_products/*.csv')
concatedData1.info()


```

#B. Fitur Transformation

## 1.Hapus Kolom Unnamed : 0

```python
concatedData1.drop('Unnamed: 0',axis=1,inplace=True) #menghapus kolom unnamed: 0 karena tdk diperlukan
concatedData1.head()
```

##2.Find and handle null Values

```python
concatedData1.isna().sum() #cek apakah ada nilai null.
```

dari hasil di atas, dapat diketahui yg terdapat null values adalah kolom ratings, no_of_ratings dan kolom discount_price, actual_price. untuk yg terkait ratings bisa dibiarkan saja, untuk yg terkait price, perlu diisi discount_price disamakan dg actual_price. Karena saya anggap tidak ada diskon.

```python
concatedData1['discount_price'] = concatedData1['discount_price'].fillna(concatedData1['actual_price'])
concatedData1.isna().sum()
```

setelah data discount_price yg null diisi dg value dari actual_price, dapat dilihat di atas bahwa jumlah row discunt_price yg null menurun drastis, dan sama dg actual_price

##3.Cek Datatype dan Ubah Ke yang lebih sesuai

###i. Ubah Rating ke float datatype

```python
#karena ada error saat ubah data langsung ke float, mengindikasikan ada data selain numeric
#maka saya ubah dulu ke numeric datanya, dan yang error diubah jadi NaN (null) values sbb:
concatedData1['ratings'] = pd.to_numeric(concatedData1['ratings'], errors='coerce')

#kemudian saya ubah menjadi tipe data float
concatedData1['ratings'] = concatedData1['ratings'].astype(float)
concatedData1.info()
```

```python
concatedData1.head()
```

###ii. Ubah no_of_ratings ke float datatype

```python
#saya hilangkan dulu nilai koma di data no_of_ratings
concatedData1['no_of_ratings'] = concatedData1['no_of_ratings'].str.replace(',', '')
concatedData1.head()
```

```python
#karena ada error saat ubah data langsung ke float, mengindikasikan ada data selain numeric
#maka saya ubah dulu ke numeric datanya, dan yang error diubah jadi NaN (null) values sbb:
concatedData1['no_of_ratings'] = pd.to_numeric(concatedData1['no_of_ratings'], errors='coerce')
```

```python
#kemudian saya ubah tipe data ke float
concatedData1['no_of_ratings'] = concatedData1['no_of_ratings'].astype(float)
concatedData1.info()
```

###iii. Ubah discount_price ke float datatype

```python
#cek apakah semua discount price dalam currency yg sama
concatedData1['discount_price'].str[0].unique()
```

dari hasil di atas, currency nya sama semua yaitu ₹, sehingga data bisa diolah atau dilakukan proses aggregate\

```python
#saya hilangkan dulu nilai koma jika ada di data discount_price
concatedData1['discount_price'] = concatedData1['discount_price'].str.replace(',', '')
concatedData1.head()
```

```python
#extract nilai numeric saja di kolom discount_price
concatedData1['discount_price'] = concatedData1['discount_price'].str.extract(r'(\d+.\d+)')[0]
```

```python
#ubah datatype
concatedData1['discount_price'] = concatedData1['discount_price'].astype(float)
concatedData1.info()
```

###iv. Ubah actual_price ke float datatype

```python
#cek apakah semua actual_price dalam currency yg sama
concatedData1['actual_price'].str[0].unique()
```

dari hasil di atas, currency nya sama semua yaitu ₹, sehingga data bisa diolah atau dilakukan proses aggregate

```python
#saya hilangkan dulu nilai koma jika ada di data actual_price
concatedData1['actual_price'] = concatedData1['actual_price'].str.replace(',', '')
concatedData1.head()
```

```python
#extract nilai numeric saja di kolom actual_price
concatedData1['actual_price'] = concatedData1['actual_price'].str.extract(r'(\d+.\d+)')[0]
```

```python
#ubah datatype actual_price ke float
concatedData1['actual_price'] = concatedData1['actual_price'].astype(float)
concatedData1.info()
```

##4.Tambah Kolom discount_rate (hasil aggregate melibatkan discount_price dan actual_price)

```python
concatedData1['discount_rate'] = concatedData1['discount_price'] / concatedData1['actual_price']
concatedData1.head()
```
